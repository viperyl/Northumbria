## Import lib

```python
import numpy as np
import matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from filterpy.kalman import KalmanFilter
warnings.simplefilter('ignore')
matplotlib.rcParams['figure.dpi'] = 100
sns.set()
%matplotlib inline
```

## Load Data

```python
file_path = 'C:\\Users\\w18029876\\Downloads\\6000-report-2\\Data\\Bollard\\SensorTile_Log_N001.txt'
df = pd.read_csv(file_path, delimiter = '\t', index_col = False)

fig, axes = plt.subplots(3, 2, figsize = (14, 27), dpi = 100)
ax_index = [0, 1, 2] * 2
df = df.drop(np.linspace(134250, len(df)-1, len(df)-134250).astype('int').tolist())
df['AccZ [mg]'] = df['AccZ [mg]'] - 1000


for i in range(2, 8):
    df[df.columns[i]].plot(ax = axes[ax_index[i-2]][int(i>=5)],alpha=0.8, color='tab:blue').set_title(df.columns[i])
    axes[ax_index[i-2]][int(i>=5)].set_title(df.columns[i], fontsize = 13)
    plt.subplots_adjust(hspace = 0.45)
```
